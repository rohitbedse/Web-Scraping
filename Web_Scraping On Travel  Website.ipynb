```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.trustpilot.com/categories/travel_insurance_company',headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
company = soup.find_all('a',class_='link_internal__7XN06 link_wrapper__5ZJEx styles_linkWrapper__UWs5j')
```

```python
name = []
website = []
rating = []
img_link = []

for i in company:

  name.append(i.find('p').text.strip())
  website.append(i.find('p',class_='typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l styles_websiteUrlDisplayed__QqkCT').text.strip())
  rating.append(i.find('span',class_='typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l styles_trustScore__8emxJ').text.replace('TrustScore ', '').strip())
  img_link.append(i.find('img').get('src') if i.find('img') else None)

```

```python
d={'Name':name,'Website':website,'Rating':rating,'Image':img_link}

df=pd.DataFrame(d)

df
```
